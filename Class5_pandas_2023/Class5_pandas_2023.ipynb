### Missing data

```python
df_temp = pd.read_csv(
    '/Users/internmet/Desktop/nine_vscode/AI_med/Class5_pandas_2023/sheets/data_research_clean/test_miss_data.csv')
print('\n-------- data --------')
print(df_temp)

## check missing data
print('\n-------- Check each value whereas Na or not --------')
print(df_temp.isna())  ## or data.isnull() --> same method

print('\n-------- Check Na but be inverse of isna --------')
print(df_temp.notna())  ## True if value exit

print('\n-------- Sum missed data each column --------')
print(df_temp.isna().sum())

print('\n-------- Sum all missing data --------')
print(df_temp.isna().sum().sum())
```

```python

## Dropna

print(df_temp)

print('\n-------- drop rows --------')
print(df_temp.dropna())

print('\n-------- drop row with particular col --------')
print(df_temp.dropna(subset=['value1']))

print('\n-------- drop col --------')
print(df_temp.dropna(axis=1))
```

```python
## Fillna
print(df_temp)

print('\n-------- Fillna with specific value --------')
print(df_temp.fillna('xxx'))

print('\n-------- Fillna with specific col and specific value --------')
fillna_dict = {'value': 114, 'value1': 'asa'}
print(df_temp.fillna(value=fillna_dict))

print('\n-------- Fillna with specific col by its mean value --------')
df_new1 = df_temp.copy()
df_new1['num_miss'] = df_new1['num_miss'].fillna(df_new1['num_miss'].mean())
print(df_new1)

print('\n-------- Fillna with specific col by backward filling--------')
df_new1 = df_temp.copy()
df_new1['num_miss'] = df_new1['num_miss'].bfill()
print(df_new1)

print('\n-------- Fillna with specific col by forward filling --------')
df_new1 = df_temp.copy()
df_new1['num_miss'] = df_new1['num_miss'].ffill()
print(df_new1)
```

```python
## Replace
import numpy as np
print('\n-------- Replace specific value with NA--------')
df_new1 = df_temp.copy()
df_new1['999_is_NA'] = df_new1['999_is_NA'].astype('str').replace('999', np.nan)
print(df_new1)
```

```python
## Interpolate
print('\n-------- Interpolate value with specific col --------')
df_new1 = df_temp.copy()
df_new1['num_miss'] = df_new1['num_miss'].interpolate(
    limit_direction='both')  ## fill with mean of next and previous value
print(df_new1)
```

### Data types

```python
df_type = pd.read_csv('/Users/internmet/Desktop/nine_vscode/AI_med/Class5_pandas_2023/sheets/data_research_clean/test_data_type.csv')
print('\n-------- data --------')
print(df_type)

## Observe data's type
print('\n-------- Observe data\'s type  --------')
print(df_type.dtypes)
print(df_type.info())

## Convert data's type
print('\n-------- Convert data\'s type (astype) --------')
df_type['cat_or_dog'] = df_type['cat_or_dog'].astype('category')
print(df_type.info())

print('\n-------- Convert data\'s type (to_numeric) --------')
df_type['str+int+float'] = pd.to_numeric(df_type['str+int+float'], errors='coerce').fillna(0)
print(df_type.info())
print(df_type)

## Convert datetime
print('\n-------- Convert data\'s type (datetime) --------')
df_type['test_date1'] = pd.to_datetime(df_type['test_date1'], )
df_type['test_date2'] = pd.to_datetime(df_type['test_date2'], dayfirst=True)
df_type['test_date3'] = pd.to_datetime(df_type['test_date3'], format='%d/%m/%Y')
df_type['test_date4'] = df_type['test_date4'].apply(lambda x: str(int(x[:4]) - 543) + x[4:])
df_type['test_date4'] = pd.to_datetime(df_type['test_date4'], format='%Y**%d**%m')
df_type['real_datetime_type'] = pd.to_datetime(df_type['real_datetime_type'], format='%Y-%m-%d %H:%M:%S%z')
df_type['real_but_no_time_zone'] = pd.to_datetime(df_type['real_but_no_time_zone'], format='%m/%d/%Y %H:%M')

df_type['year_real'] = df_type['real_datetime_type'].dt.year
df_type['month_real'] = df_type['real_datetime_type'].dt.month
df_type['day_real'] = df_type['real_datetime_type'].dt.day

print(df_type.to_string())
print(df_type.info())
```

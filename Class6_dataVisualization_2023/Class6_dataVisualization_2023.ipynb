### Multi axes

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(2 * x)

plt.subplot(311)
plt.plot(x, y1, label='sin(x)', color='red')

plt.subplot(312)
plt.plot(x, y2, label='cos(x)', color='blue')
# plt.grid()

plt.subplot(313)
plt.scatter(x,y3)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(2 * x)

plt.figure(figsize=(12,6))
plt.suptitle('Testing_multi-axes_plotting',fontsize=15)

plt.subplot(311)
plt.plot(x, y1, label='sin(x)', color='red')
plt.ylabel('Amplitude(mV)')
plt.grid()

plt.subplot(312)
plt.title('Here is title of second axes')
plt.plot(x, y2, label='cos(x)', color='blue')

plt.subplot(313)
plt.scatter(x,y3)
plt.xlim(-5,5)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(2 * x)

fig, axs = plt.subplots(3, 1, figsize=(12,6))
fig.suptitle('Testing_multi-axes_plotting',fontsize=15)

# First plot
axs[0].plot(x, y1, label='sin(x)', color='red')
axs[0].set_ylabel('Amplitude(mV)')
axs[0].grid()

# Second plot
axs[1].set_title('Here is title of second axes')
axs[1].plot(x, y2, label='cos(x)', color='blue')

# Third plot
axs[2].scatter(x, y3)
axs[2].set_xlim(-5,5)

plt.tight_layout()
plt.show()
```

```python
import plotly.graph_objects as go

# Load the data
df = pd.read_csv('sheets/worldometer_data_original.csv')

# Calculate the ratio of Total Deaths per Total Cases
df['Death_Case_Ratio'] = df['TotalDeaths'] / df['TotalCases']

# Drop rows with missing values in relevant columns
df = df.dropna(subset=['TotalDeaths', 'Country/Region', 'Continent', 'Death_Case_Ratio'])

# Sort values based on Death to Case ratio
df = df.sort_values(['Death_Case_Ratio'], ascending=False)

# Creating the main figure
fig = go.Figure()

# Creating a list of continents
continents = df['Continent'].unique()

# Adding data to the figure for each continent
for continent in continents:
    subset = df[df['Continent'] == continent]
    fig.add_trace(
        go.Bar(
            x=subset.head(20)['Country/Region'],
            y=subset.head(20)['Death_Case_Ratio'],
            name=continent
        )
    )

# Updating the layout to create the dropdown menu
fig.update_layout(
    updatemenus=[
        dict(
            active=0,
            buttons=[
                dict(label=continent,
                     method="update",
                     args=[{"visible": [continent == c for c in continents]},
                           {"title": f"Top 20 Countries in {continent} by Death to Case Ratio"}])
                for continent in continents
            ]
        )
    ]
)

# Updating the layout to adjust the figure size and add axis labels
fig.update_layout(
    title_text="Top 20 Countries by Death to Case Ratio",
    xaxis_title="Country/Region",
    yaxis_title="Death to Case Ratio",
    width=1200,
    height=800
)

# Setting the first continent to be visible by default
fig.data[0].visible = True

# Show the plot
fig.show()
```

```python
import matplotlib.pyplot as plt

df = pd.read_csv('sheets/worldometer_data_original.csv')

# Calculate the Case Fatality Rate (CFR) for each country
df['CFR'] = (df['TotalDeaths'] / df['TotalCases']) * 100

# Group by Continent, summing the total cases and total deaths, and then calculate CFR
continent_cfr = df.groupby('Continent').agg({'TotalCases': 'sum', 'TotalDeaths': 'sum'})
continent_cfr['CFR'] = (continent_cfr['TotalDeaths'] / continent_cfr['TotalCases']) * 100

# Sort the values for better visualization
continent_cfr = continent_cfr.sort_values(by='CFR', ascending=False)

# Plotting
ax = continent_cfr['CFR'].plot(kind='bar', figsize=(10, 6), color='skyblue', edgecolor='black')

# Adding labels and title
plt.ylabel('Case Fatality Rate (%)')
plt.xlabel('Continent')
plt.title('Case Fatality Rate by Continent')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'sheets/thailand_domestic_tourism_2019_2023.csv'
tourism_data = pd.read_csv(file_path)

# Step 1: Filter the dataset
filtered_data = tourism_data[tourism_data['variable'].isin(['no_tourist_thai', 'no_tourist_foreign'])]

# Step 2: Pivot the table
pivoted_data = filtered_data.pivot_table(index=['date', 'province_thai', 'province_eng', 'region_thai', 'region_eng'], 
                                          columns='variable', 
                                          values='value').reset_index()

# Step 3: Calculate the mean for each region_eng group
mean_values = pivoted_data.groupby('region_eng').mean().reset_index()

# Step 4: Create a clustered bar chart
plt.figure(figsize=(10, 6))
sns.barplot(data=mean_values.melt(id_vars='region_eng'), x='region_eng', y='value', hue='variable')
plt.title('Mean Number of Tourists by Region (Thai and Foreign)')
plt.ylabel('Mean Number of Tourists')
plt.xlabel('Region')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'sheets/thailand_domestic_tourism_2019_2023.csv'
tourism_data = pd.read_csv(file_path)

# Filter the dataset
filtered_data = tourism_data[tourism_data['variable'].isin(['no_tourist_thai', 'no_tourist_foreign'])]

# Pivot the table
pivoted_data = filtered_data.pivot_table(index=['date', 'province_thai', 'province_eng', 'region_thai', 'region_eng'], 
                                          columns='variable', 
                                          values='value').reset_index()

# Select only numeric columns and 'region_eng' for grouping
numeric_data = pivoted_data.select_dtypes(include='number')
numeric_data['region_eng'] = pivoted_data['region_eng']

# Calculate the mean for each region_eng group
mean_values = numeric_data.groupby('region_eng').mean().reset_index()

# Create a clustered bar chart
plt.figure(figsize=(10, 6))
sns.barplot(data=mean_values.melt(id_vars='region_eng'), x='region_eng', y='value', hue='variable')
plt.title('Mean Number of Tourists by Region (Thai and Foreign)')
plt.ylabel('Mean Number of Tourists')
plt.xlabel('Region')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
